```python
import pandas as pd
```

```python
raw_path = "/kaggle/input/persian-news-dataset/archive_v5.csv"
df_raw = pd.read_csv(raw_path)

df_raw.head()
```

```python
df_raw.isna().sum()
```

```python
df = df_raw.copy()

# Remove rows with missing 'body' values
df = df.dropna(subset='body')

df.isna().sum()
```

```python
df[df['body']=='body']
```

```python
# Remove rows with 'body' value equal to 'body'
df = df.drop(index=df[df['body']=='body'].index)
df = df.reset_index(drop=True)

df[df['body']=='body']
```

```python
df[df['body']==' '*5]
```

```python
# Remove rows with 'body' value equal to just some space characters: ' '*n
idx = []
for i in range(len(df)):
    if df['body'].iloc[i].strip() == '':
        idx.append(i)

df = df.drop(index=idx)
df = df.reset_index(drop=True)

df[df['body']==' '*5]
```

```python
# Counting news with less than 400 characters
news_len = [len(news) for news in df['body']]
short = [a for a in news_len if a<400]
len(short)
```

```python
# Removing short news
idx = [i for i in range(len(df)) if len(df['body'].iloc[i])<400]
df = df.drop(index=idx)
df = df.reset_index(drop=True)
```

```python
# Sampling 20,000 random news for the subset dataset
df_subset = df.sample(n=20000)
df_subset = df_subset.reset_index(drop=True)
df_subset
```

```python
import re

def text_normalizer(text):
    """
    Normalize Persian text by:
    - Converting Arabic characters to Persian equivalents
    - Normalizing digits to Persian digits
    - Removing diacritics
    - Normalizing whitespace and zero-width characters
    """

    if not isinstance(text, str):
        raise TypeError("Input must be a string")

    # Arabic → Persian character mapping
    char_map = {
        "ي": "ی",
        "ك": "ک",
        "ى": "ی",
        "ئ": "ی",
        "ؤ": "و",
        "ة": "ه",
        "ۀ": "ه",
        "إ": "ا",
        "أ": "ا",
        "آ": "آ",
        "ٱ": "ا",
        "ـ": "",
    }

    for arabic_char, persian_char in char_map.items():
        text = text.replace(arabic_char, persian_char)

    # Remove Arabic diacritics
    diacritics_pattern = re.compile(
        r"[\u064B-\u065F\u0670\u06D6-\u06ED]"
    )
    text = diacritics_pattern.sub("", text)

    # Normalize English digits → Persian digits
    english_digits = "0123456789"
    persian_digits = "۰۱۲۳۴۵۶۷۸۹"
    digit_map = str.maketrans(english_digits, persian_digits)
    text = text.translate(digit_map)

    # Remove zero-width characters
    zero_width_chars = [
        "\u200c",
        "\u200d",
        "\u200e",
        "\u200f",
        "\ufeff",
    ]
    for ch in zero_width_chars:
        text = text.replace(ch, " ")

    # Normalize whitespace
    text = re.sub(r"\s+", " ", text).strip()

    return text

```

```python
df_subset['body'] = df_subset['body'].apply(text_normalizer)
```

```python
# Saving the dataset
df_subset.to_csv('news_subset.csv', index=False)
```
